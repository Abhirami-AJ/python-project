```python
import numpy as np
import pandas as pd


df = pd.read_csv("1774404110.csv")


df.columns = df.columns.str.strip().str.lower()


print("Columns:", df.columns)



mpg = df['mpg'].values
cylinders = df['cylinders'].values
displacement = df['displacement'].values
origin = df['origin'].values


horsepower = pd.to_numeric(df['horsepower'], errors='coerce').values


car_name = df['car name'].values


print("\n--- Basic Array Operations ---")
print("Mean MPG:", np.mean(mpg))
print("Median MPG:", np.median(mpg))
print("Std Dev MPG:", np.std(mpg))
print("Cars with MPG > 25:", np.sum(mpg > 25))


print("\n--- Cars with >6 Cylinders ---")
print(list(car_name[cylinders > 6]))


mean_hp = np.nanmean(horsepower)
horsepower[np.isnan(horsepower)] = mean_hp

new_hp = horsepower * 1.10
print("\nUpdated Horsepower:", new_hp)

avg_disp_europe = np.mean(displacement[origin == 2])
print("\nAvg Displacement (Europe):", avg_disp_europe)


print("\n--- Avg MPG by Cylinders ---")
for cyl in np.unique(cylinders):
    print(f"{cyl} cylinders:", np.mean(mpg[cylinders == cyl]))




print("\nFirst 10 rows:\n", df.head(10))
print("\nShape:", df.shape)
print("\nSummary:\n", df.describe())


filtered = df[(df['model year'] == 75) & (df['weight'] < 3000)]
print("\nFiltered Data:\n", filtered[['car name', 'weight', 'mpg']])


df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)


df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
print("\nPower to Weight Ratio:\n", df[['car name', 'power_to_weight_ratio']].head())


print("\nMean MPG by Origin:\n", df.groupby('origin')['mpg'].mean())


summary = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()
print("\nSummary Data:\n", summary)
high_mpg = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg.to_csv("high_mpg_cars.csv", index=False)
```
